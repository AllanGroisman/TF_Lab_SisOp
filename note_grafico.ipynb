```python
# Se estiver no Jupyter Notebook, ative os grficos inline
%matplotlib inline

import matplotlib.pyplot as plt
import re
from collections import OrderedDict

#  Cole a sada do seu programa entre as aspas triplas abaixo
saida = """
A = 311
B = 295
C = 362
D = 327
E = 333
F = 352
G = 337
H = 220
I = 264
J = 271
"""

# Extrai as contagens por letra
contagem = OrderedDict()
for linha in saida.strip().splitlines():
    match = re.match(r'^([A-Z])\s*=\s*(\d+)', linha)
    if match:
        letra, valor = match.groups()
        contagem[letra] = int(valor)

# Se nada foi extrado, parar
if not contagem:
    print(" Nenhuma contagem encontrada.")
else:
    letras = list(contagem.keys())
    valores = list(contagem.values())
    total = sum(valores)
    percentuais = [v / total * 100 for v in valores]

    # Grfico: Frequncia absoluta
    plt.figure(figsize=(10, 4))
    plt.bar(letras, valores)
    plt.title("Frequncia absoluta por letra (execues por thread)")
    plt.xlabel("Thread (letra)")
    plt.ylabel("Quantidade")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Grfico: Frequncia relativa
    plt.figure(figsize=(10, 4))
    plt.bar(letras, percentuais)
    plt.title("Frequncia relativa por letra (%)")
    plt.xlabel("Thread (letra)")
    plt.ylabel("Porcentagem do total")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```
